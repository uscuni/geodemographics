# 03 Census data cleaning

This notebook removes unnecessary columns from the original files and creates new files with relevant information.

```python
from glob import glob
from pathlib import Path

import pandas as pd
```

```python
import os

os.getcwd()
```

```python
# Load files
files = glob("/data/uscuni-restricted/geodem/01_csv/*nadzsjd*")

# Load file with relevant column information
names = pd.read_csv(
    "/home/anna/geodemographics/notebooks/_goedem_names_gender.csv", sep=";"
)

# Columns you always want to keep
common_columns = [
    "NUTS_2",
    "naz_oblast",
    "NUTS_3",
    "naz_kraj",
    "kod_okres",
    "naz_okres",
    "kod_orp",
    "naz_orp",
    "kod_obec",
    "naz_obec",
    "kod_mco",
    "nazev_mco",
    "nadzsjd",
    "nuts_2_regs",
    "nazev_regionu_soudrznosti",
    "nuts3_kraj",
    "nazev_kraje",
    "nazev_okresu",
    "nazev_orp",
    "nazev_obec",
]

# Determine columns to drop: keep those with use==1 OR in common_columns
drop = names[names["use"] != 1]

# Remove unnecessary columns column
drop = drop.drop(columns={"use"})

# Exclude common_columns from drop list
drop_cols = [c for c in drop["full_names_CZ"] if c not in common_columns]

# Process all files
for file in files:
    path = Path(file)
    # Skip helper files that start with "_"
    if path.stem.startswith("_"):
        continue
    # Load and clean
    data = pd.read_csv(path)
    data_cleaned = data.drop(columns=drop_cols, errors="ignore")
    # Save the cleaned data
    data_cleaned.to_csv(
        f"/data/uscuni-restricted/geodem/02_preprocessed/{path.stem}_preprocessed.csv",
        index=False,
    )
```
