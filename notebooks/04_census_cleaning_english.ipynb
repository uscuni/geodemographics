# 03 Census data cleaning

This notebook removes unnecessary columns from the original files and creates new files with relevant information.

```python
from glob import glob
import os
import pandas as pd
```

```python
file_map = {
    "Byty_2021": "Apartments_General",
    "Byty_druh_obdobi_vystavby": "Apartments_Construction_Period",
    "Byty_velikost_vybaveni": "Apartments_Size_and_Equipment",
    "Byty_vytapeni": "Apartments_Heating",
    "D_Domy": "Houses_General",
    "Domacnosti": "Households",
    "ekonomicka_aktivita_pohlavi": "Economic_Activity_by_Gender",
    "nabozenstvi_pohlavi": "Religion_by_Gender",
    "pobyt_pohlavi": "Residence_Type_by_Gender",
    "rodinny_stav_pohlavi": "Marital_Status_by_Gender",
    "statni_obcanstvi_narodnost": "Citizenship_and_Nationality",
    "statni_obcanstvi": "Citizenship",
    "vek_pohlavi": "Age_and_Gender",
    "vzdelani_pohlavi": "Education_by_Gender",
    "zamestnani_ek_cinnost_pohlavi": "Employment_by_Industry_and_Gender",
    "zamestnani_postaveni_vek": "Employment_Status_by_Age",
    "zamestnani": "Employment_General",
}
```

```python
source_folder = "/data/uscuni-restricted/geodem/02_preprocessed"
output_folder = "/data/uscuni-restricted/geodem/03_cleaned"
```

```python
glob("/data/uscuni-restricted/geodem/02_preprocessed/*")
```

```python
rename_dict = {
    "nuts_2_regs": "NUTS_2",
    "nazev_regionu_soudrznosti": "naz_oblast",
    "nuts3_kraj": "NUTS_3",
    "nazev_kraje": "naz_kraj",
    "nazev_okresu": "naz_okres",
    "nazev_orp": "naz_orp",
    "nazev_obec": "naz_obec",
}
```

```python
for filename in os.listdir(source_folder):
    if filename.endswith(".csv"):
        file_path = os.path.join(source_folder, filename)
        df = pd.read_csv(file_path)

        df.rename(columns=rename_dict, inplace=True)

        for key, eng_name in file_map.items():
            if key in filename:
                new_name = f"nadzsjd_{eng_name}_2021.csv"
                break

        # 3. Save
        save_path = os.path.join(output_folder, new_name)
        df.to_csv(save_path, index=False)
        print(f"Saved: {new_name}")
```
