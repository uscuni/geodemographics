```python
geo_cols = {
    "NUTS_2",
    "naz_oblast",
    "NUTS_3",
    "naz_kraj",
    "kod_okres",
    "naz_okres",
    "kod_orp",
    "naz_orp",
    "kod_obec",
    "naz_obec",
    "kod_mco",
    "nazev_mco",
    "geometry",
}


exclude = ["nadzsjd_Citizenship_and_Nationality_2021.parquet"]
```

```python
for source_path in files:
    filename = os.path.basename(source_path)
    target_path = os.path.join("/data/uscuni-restricted/geodem/05_data", filename)

    if filename in exclude:
        print(f" Ignored: {filename}")
        continue

    if os.path.exists(target_path):
        print(f"Skipping: {filename} already exists.")
        continue

    try:
        # 2. Load and check columns
        df = gpd.read_parquet(source_path)
        current_columns = set(df.columns)
        data_columns = current_columns - geo_cols

        if not data_columns:
            print(f"Ignored: {filename} has no data columns.")
        else:
            # 3. Save copy
            df.to_parquet(target_path)
            print(f"Moved: {filename}")

    except Exception as e:
        print(f"Error with {filename}: {e}")
```
